```python
fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot([0.0, 0, 1.0, 1.0])
ax1 = plt.plot(np.arange(0, 5, 0.2))

f, ax = plt.subplots(2, 2) #количество координатных осей

f.set_size_inches(10, 4)     #размер 7 х 4 дюйма
f.set_facecolor('#eee')     #цвет фона (светло-серый)

ax[0, 0].plot(np.arange(0, 5, 0.2))
ax[0, 0].grid()

ax[0, 1].plot(np.arange(5, 0, -0.2))
ax[0, 1].grid()

plt.show()
```

```python
from matplotlib.gridspec import GridSpec
```

```python
ws = [1, 2, 5]      #размер полей отдельных графиков
hs = [2, 0.5]       #размер полей отдельных графиков

fig = plt.figure(figsize=(7, 4))
gs = GridSpec(ncols=3, nrows=2, figure=fig, width_ratios=ws, height_ratios=hs)

ax1 = plt.subplot(gs[0, 0])
ax1.plot(np.arange(0, 5, 0.2))
ax2 = fig.add_subplot(gs[1, 0:2])
ax2.plot(np.random.random(10))
ax3 = fig.add_subplot(gs[:, 2])
ax3.plot(np.random.random(10))

plt.show()
```
