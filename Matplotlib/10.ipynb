# 10. Рисуем гистограммы, столбчатые и круговые диаграммы

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Гистограммы

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
ax.hist(y)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
ax.hist(y, 50)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
x = np.linspace(np.min(y), np.max(y), 10)
bars = [len(y[np.bitwise_and(y >= x[i], y < x[i+1])]) for i in range(len(x)-1)]

#x = [f'H{i+1}' for i in range(10)]
#y = np.random.randint(1, 5, len(x))

ax.bar(range(len(x) - 1), bars)

#ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(1, 5, len(x))

ax.bar(x, y)

#ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

y = np.random.normal(0, 2, 500)
x = np.linspace(np.min(y), np.max(y), 10)
bars = [len(y[np.bitwise_and(y >= x[i], y < x[i+1])]) for i in range(len(x)-1)]

#x = [f'H{i+1}' for i in range(10)]
#y = np.random.randint(1, 5, len(x))

ax.barh(range(len(x) - 1), bars) #горизонтальное отображение

#ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(-20, 20, len(x))

ax.bar(x, y, width=0.5, linewidth=2, edgecolor='r', yerr=2, bottom=10)

#ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

x = np.arange(10)
y1 = np.random.randint(3, 20, len(x))
y2 = np.random.randint(3, 20, len(x))
w = 0.3
ax.bar(x - w/2, y1, width=w)
ax.bar(x + w/2, y1, width=w)

ax.grid()

plt.show()
```
