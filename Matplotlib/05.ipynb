# 5. Настраиваем формат отображения меток у координатных осей

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter, FormatStrFormatter
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x))

ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x))

ax.set_xticklabels([])
ax.set_yticklabels([])

ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x))

ax.xaxis.set_major_formatter(NullFormatter())

ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x))

ax.yaxis.set_major_formatter(FormatStrFormatter("%d"))  #отображение только целых чисел

ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x))

ax.yaxis.set_major_formatter(FormatStrFormatter("y=%.2f"))  #отображение вещественные значения

ax.grid()
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

def format0y(x, pos):
    return f"[{x}]" if x < 0 else f"({x})"

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x))

ax.yaxis.set_major_formatter(FuncFormatter(format0y))  

ax.grid()
plt.show()
```

```python
from matplotlib.ticker import ScalarFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x) * 1e5)

ax.yaxis.set_major_formatter(ScalarFormatter())  

ax.grid()
plt.show()
```

```python
from matplotlib.ticker import ScalarFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x) * 1e10)

ax.yaxis.set_major_formatter(ScalarFormatter())  

ax.grid()
plt.show()
```

```python
from matplotlib.ticker import ScalarFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x) * 1e5)

sf = ScalarFormatter()
sf.set_powerlimits((-4, 4)) #определение границ степени для форматора
ax.yaxis.set_major_formatter(sf)  

ax.grid()
plt.show()
```

```python
import matplotlib
matplotlib.rcParams['axes.formatter.limits'] = (-4, 4) #вместо форматора

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x) * 1e5)

#sf = ScalarFormatter()
#sf.set_powerlimits((-4, 4)) #определение границ степени для форматора
#ax.yaxis.set_major_formatter(sf)  

ax.grid()
plt.show()
```

```python
from matplotlib.ticker import FixedFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)

ax.plot(x, np.sin(x))

ax.xaxis.set_major_formatter(FixedFormatter([-3, -2, -1, 0, 1, 2, 3]))  

ax.grid()
plt.show()
```
