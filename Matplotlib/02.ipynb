```python
y = np.arange(0, 5, 1)              # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y ])     # [0, 1, 4, 9, 16]

y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]

plt.plot(x, y, x2, y2)
plt.grid()
plt.show()
```

```python
y = np.arange(0, 5, 1)              # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y ])     # [0, 1, 4, 9, 16]

y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]

plt.plot(x, y)
plt.plot(x2, y2)
plt.grid()
plt.show()
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y)

plt.grid()
plt.show()
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y, 0.5, where=(y < 0))

plt.grid()
plt.show()
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y, 0.5, where=(y < 0), color='r', alpha=0.4)
plt.fill_between(x, y, 0.5, where=(y > 0), color='g', alpha=0.3)

plt.grid()
plt.show()
```
