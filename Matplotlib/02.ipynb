```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.array([1, 2, -6, 0, 4])
```

```python
plt.plot(y)
plt.show()
```

```python
x = np.array([4, 5, 6, 7, 8])
y = np.array([1, 2, -6, 0, 4])

plt.plot(x, y)
plt.show()
```

```python
x = np.array([1, 1, 5, 5, 1])
y = np.array([1, 5, 5, 1, 1])

plt.plot(x, y)
plt.show()
```

```python
y = np.arange(0, 5, 1)              # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y ])     # [0, 1, 4, 9, 16]

y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]

plt.plot(x, y, x2, y2)
plt.grid()
plt.show()
```
