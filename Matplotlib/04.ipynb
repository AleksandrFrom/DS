```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator, LinearLocator
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set(xlim=(-5, 30), ylim=(-1, 6)) #граничные значения осей х и у

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set_xlim(-5, 30)
ax.set_ylim(-1, 6)

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set_xlim(xmin=-5, xmax=10)   #ограничение графика по оси х
ax.set_ylim(ymin=-1, ymax=6)    #ограничение графика по оси у

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

lc = NullLocator()

ax.grid()
ax.xaxis.set_major_locator(lc) #сокрытие меток по выбранной оси

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

lc = NullLocator()

ax.grid()
ax.yaxis.set_major_locator(lc) #сокрытие меток по выбранной оси

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.grid()
ax.yaxis.set_major_locator(LinearLocator(5)) 

plt.show()
```

```python
from matplotlib.ticker import MultipleLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))

ax.grid()
ax.xaxis.set_major_locator(MultipleLocator(base=1)) 
ax.yaxis.set_major_locator(MultipleLocator(base=0.5)) 

plt.show()
```

```python
from matplotlib.ticker import IndexLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.xaxis.set_major_locator(IndexLocator(base=0.5, offset=0)) 

plt.show()
```

```python
from matplotlib.ticker import IndexLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.xaxis.set_major_locator(IndexLocator(base=0.5, offset=0.57)) 

plt.show()
```

```python
from matplotlib.ticker import FixedLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.xaxis.set_major_locator(FixedLocator([-2, -1, 0, 1, 2, 3])) 

plt.show()
```

```python
from matplotlib.ticker import LogLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.xaxis.set_major_locator(LogLocator(base=2)) 

plt.show()
```

```python
from matplotlib.ticker import MaxNLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.xaxis.set_major_locator(MaxNLocator(5)) 

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.minorticks_on()
ax.grid(which='major', lw = 2)
ax.grid(which='minor')

ax.xaxis.set_major_locator(MaxNLocator(5)) 

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.minorticks_on()
ax.grid(which='major', lw = 2)
ax.grid(which='minor')

ax.xaxis.set_minor_locator(NullLocator()) 

plt.show()
```
