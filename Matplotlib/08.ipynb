# 8. Добавляем легенду и рисуем геометрические фигуры на графиках

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

ax.plot(np.arange(0, 5, 0.25), '--o' ,label='line 1')
ax.plot(np.arange(0, 10, 0.5), ':s' ,label='line 2')

ax.legend(['1', '2']) #отображение легендны на графике

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

ax.plot(np.arange(0, 5, 0.25), '--o')
ax.plot(np.arange(0, 10, 0.5), ':s' )

ax.legend(['1', '2']) #отображение легендны на графике

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s' )

ax.legend((line2, line1), ['Линия 2', 'Линия 1']) #отображение легендны на графике

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s' )

ax.legend((line2, line1), ['Линия 2', 'Линия 1'], loc='upper right') #отображение в верхнем правом углу

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s' )

ax.legend((line2, line1), ['Линия 2', 'Линия 1'], bbox_to_anchor=(0.5, 0.7)) #отображение легенды по координатам

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s' )

ax.legend((line2, line1), [r'$f(x) = a\cdot b + c$', r'$f(x)=k \cdot x + b$']) #отображение формул

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s' )

ax.legend((line2, line1), [r'$f(x) = a\cdot b + c$', r'$f(x)=k \cdot x + b$'],
          facecolor='#aaa', framealpha = 0.2) #отображение формул

plt.show()
```

Рисование ломаной линии

```python
from matplotlib.lines import Line2D

l1 = Line2D([1, 2, 3], [1, 2, 4])

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

line1, = ax.plot(np.arange(0, 5, 0.25), '--o')
line2, = ax.plot(np.arange(0, 10, 0.5), ':s' )

ax.add_line(l1)

plt.show()
```

```python
l1 = Line2D([1, 2, 3], [1, 2, 4])

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

ax.add_line(l1)
ax.set(xlim=(1, 3), ylim=(1, 4))

plt.show()
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
cos = Line2D(x, np.cos(x))

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

ax.add_line(cos)
ax.set(xlim=(-2*np.pi, 2*np.pi), ylim=(-1.2, 1.2))

plt.show()
```

```python
from matplotlib.patches import *

x = np.arange(-2*np.pi, 2*np.pi, 0.1)
cos = Line2D(x, np.cos(x))

rect = Rectangle((0, 0), 2.5, 0.5, facecolor='g')

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

ax.add_line(cos)
ax.add_patch(rect)
ax.set(xlim=(-2*np.pi, 2*np.pi), ylim=(-1.2, 1.2))

plt.show()
```
