# 6. Делаем логарифмический масштаб у координатных осей

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.semilogy(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.semilogx(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.set_yscale('log') #масштаб оси
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.set_yscale('log', base=2)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.set_yscale('log', base=10, subs=[2, 9]) #показать риску 2 и 9
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.1)
ax.plot(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.set_xscale('symlog', linthresh=2, linescale=5)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.1)
ax.plot(x, np.sinc(x) * np.exp(-np.abs(x/10)))
ax.set_xscale('symlog', linthresh=2)
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10*np.pi, 10*np.pi, 0.1)
ax.loglog(x, np.sinc(x) * np.exp(-np.abs(x/10))) #логарифмические оси по х и у

ax.grid()

plt.show()
```
