# 9. Рисуем ступенчатые, стековые, stem и точечные графики

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ступенчатый график

fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

x = np.arange(0, 10)
ax.step(x, x, '-go') #ступенчатый график
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

x = np.arange(0, 10)
ax.step(x, x, '-go', where='post') #ступенчатый график
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

x = np.arange(0, 10)
ax.step(x, x, '-go', where='mid') #ступенчатый график
ax.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

x = np.arange(0, 10)
ax.step(x, x, '-go', np.cos(x), '--x', where='post') #ступенчатый график
ax.grid()

plt.show()
```

```python
#стековый график

fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

x = np.arange(-2, 2, 0.1)
y1 = np.array([-y**2 for y in x]) + 8
y2 = np.array([-y**2 for y in x]) + 8
y3 = np.array([-y**2 for y in x]) + 8
ax.stackplot(x, y1, y2, y3)

ax.grid()

plt.show()
```

```python
#ste-график

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

x = np.arange(-np.pi, np.pi, 0.3)
ax.stem(x, np.cos(x), bottom=0.5)

ax.grid()

plt.show()
```

```python
#ste-график

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

x = np.arange(-np.pi, np.pi, 0.3)
ax.stem(x, np.cos(x), '--r', bottom=0.5)

ax.grid()

plt.show()
```

```python
#точечный график

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()

x = np.random.normal(0, 2, 500)
y = np.random.normal(0, 2, 500)

ax.scatter(x, y, s=50, c='g', linewidths=1, marker='s', edgecolors='r')

ax.grid()

plt.show()
```
